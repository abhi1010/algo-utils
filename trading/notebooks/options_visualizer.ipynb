```python
import volvisualizer.volatility as vol

imp = vol.Volatility(ticker='SPY')

# Ref: https://github.com/GBERESEARCH/volvisualizer/blob/master/notebooks/Implied%20Volatility%20-%20SPX%20Walkthrough.ipynb

;
```

```python
# imp.surface()


def plot_fn(plot_type):

  if plot_type == 1:
    imp.visualize(
        graphtype='surface', surfacetype='spline', scatter=True, smoothing=True)
  elif plot_type == 2:
    imp.visualize(
        graphtype='surface', surfacetype='interactive_spline', smoothing=True)
  elif plot_type == 3:
    imp.visualize(
        graphtype='surface',
        surfacetype='interactive_spline',
        smoothing=True,
        colorscale='Blues',
        scatter=True,
        opacity=0.8)
  elif plot_type == 4:
    imp.visualize(
        graphtype='surface',
        surfacetype='interactive_spline',
        rbffunc='cubic',
        colorscale='Jet',
        smoothing=True)
  elif plot_type == 5:
    imp.visualize(
        graphtype='surface',
        surfacetype='interactive_spline',
        rbffunc='cubic',
        colorscale='Jet',
        smoothing=True)
```

```python
plot_fn(1)
```

```python

plot_fn(5)
```

```python
plot_fn(2)
```
